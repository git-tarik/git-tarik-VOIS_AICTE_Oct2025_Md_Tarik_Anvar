```python
# Step 1: Import libraries
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset (make sure the file is uploaded in Colab)
df = pd.read_csv("space_missions.csv",encoding='latin1')

# Display first 5 rows
df.head()
```

```python
# Step 2: Convert Date column to datetime format
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')

# Extract Year and Decade
df['Year'] = df['Date'].dt.year
df['Decade'] = (df['Year'] // 10) * 10   # e.g., 1957 → 1950s

# Check if conversion worked
df[['Date', 'Year', 'Decade']].head(10)
```

```python
# Step 3: Number of missions per year
missions_per_year = df.groupby('Year').size()

plt.figure(figsize=(12,6))
missions_per_year.plot(kind='line', marker='o')
plt.title("Number of Space Missions per Year")
plt.xlabel("Year")
plt.ylabel("Number of Missions")
plt.grid(True)
plt.show()
```

```python
# Step 4: Number of missions per decade
missions_per_decade = df.groupby('Decade').size()

plt.figure(figsize=(10,6))
missions_per_decade.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title("Number of Space Missions per Decade")
plt.xlabel("Decade")
plt.ylabel("Number of Missions")
plt.xticks(rotation=45)
plt.show()
```
