```python
df['Price']=df['Price'].fillna(df['Price'].median())
```

```python
df.isnull().sum()
```

```python
#Calculate Skewness
skew_value = df['Price'].skew()
print("Skewness:", skew_value)

#Plot Histogram
plt.hist(df['Price'], bins=10,color='skyblue',edgecolor='black')
plt.title('Histogram of Price (Skewness={skew_value:.2f})')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()
```

```python
#Duplicate values
duplicates=df[df.duplicated()]
print('Number of duplicate rows:',len(duplicates))
```

```python
start =len(df)
df.drop_duplicates(inplace=True)
end=len(df)
print('Number of duplicate rows removed:',start-end)
```

```python
#Outliers
plt.boxplot(df['Price'])
plt.show()
```

```python
#Remove Outliers
def remove_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    #cap the outliers
    df[column] = df[column].apply(lambda x:lower_bound if x<lower_bound else upper_bound if x>upper_bound else x)
    return df
#update only price column in the original Dataframe
df= remove_outliers(df,'Price')
```
